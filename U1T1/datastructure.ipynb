# Explicação do Exercício 2.1

## Contexto do Exercício

O exercício envolve calcular as probabilidades de todos os resultados possíveis ao lançar duas moedas que têm probabilidades diferentes de cair "cara" (H) e "coroa" (T). As combinações possíveis de resultados ao lançar duas moedas são:

1. Cara - Cara (H-H)
2. Cara - Coroa (H-T)
3. Coroa - Cara (T-H)
4. Coroa - Coroa (T-T)

## Probabilidades Definidas

Para resolver este problema, consideramos as seguintes probabilidades para cada moeda:

- Para a **moeda 1**:
  - \( p_1(H) \): Probabilidade de sair cara.
  - \( p_1(T) = 1 - p_1(H) \): Probabilidade de sair coroa.

- Para a **moeda 2**:
  - \( p_2(H) \): Probabilidade de sair cara.
  - \( p_2(T) = 1 - p_2(H) \): Probabilidade de sair coroa.

## Cálculo das Probabilidades

Usamos as regras de probabilidade para calcular as chances de cada resultado:

1. **Probabilidade de Cara - Cara (H-H)**:
   - Esta probabilidade é calculada multiplicando a probabilidade de obter cara na moeda 1 pela probabilidade de obter cara na moeda 2:
   \[
   P(H-H) = p_1(H) \times p_2(H)
   \]

2. **Probabilidade de Cara - Coroa (H-T)**:
   - Esta probabilidade é obtida multiplicando a probabilidade de obter cara na moeda 1 pela probabilidade de obter coroa na moeda 2:
   \[
   P(H-T) = p_1(H) \times p_2(T) = p_1(H) \times (1 - p_2(H))
   \]

3. **Probabilidade de Coroa - Cara (T-H)**:
   - Esta probabilidade é calculada multiplicando a probabilidade de obter coroa na moeda 1 pela probabilidade de obter cara na moeda 2:
   \[
   P(T-H) = p_1(T) \times p_2(H) = (1 - p_1(H)) \times p_2(H)
   \]

4. **Probabilidade de Coroa - Coroa (T-T)**:
   - Esta probabilidade é obtida multiplicando a probabilidade de obter coroa na moeda 1 pela probabilidade de obter coroa na moeda 2:
   \[
   P(T-T) = p_1(T) \times p_2(T) = (1 - p_1(H)) \times (1 - p_2(H))
   \]

## Resultado

Após calcular essas probabilidades, podemos determinar a distribuição dos resultados possíveis ao lançar as duas moedas. Esse exercício utiliza conceitos fundamentais de probabilidade e combinações de eventos independentes, permitindo calcular a probabilidade de cada resultado possível considerando as características específicas das moedas (carregadas).


````python
import numpy as np
import pandas as pd

# Definindo as probabilidades carregadas para cada moeda (p1, p2)
probabilities = {
    'p1(H)': [0.5, 0.6, 0.4, 0.1, 0.3],
    'p2(H)': [0.5, 0.7, 0.8, 0.2, 0.4]
}

# Criar um DataFrame para armazenar os valores
df = pd.DataFrame(probabilities)

# Calcular as probabilidades para os resultados (H-H, H-T, T-H, T-T)
df['H-H'] = df['p1(H)'] * df['p2(H)']  # Probabilidade de Cara-Cara
df['H-T'] = df['p1(H)'] * (1 - df['p2(H)'])  # Probabilidade de Cara-Coroa
df['T-H'] = (1 - df['p1(H)']) * df['p2(H)']  # Probabilidade de Coroa-Cara
df['T-T'] = (1 - df['p1(H)']) * (1 - df['p2(H)'])  # Probabilidade de Coroa-Coroa

# Exibir os resultados
print(df)
````

# Explicação do Exercício 2.2

## Contexto do Exercício

O exercício envolve a aplicação do Teorema de Bayes para calcular a probabilidade de um e-mail ser classificado como spam, dado que uma palavra suspeita foi encontrada. Essa situação é comum em filtros de spam, onde o objetivo é aumentar a precisão na identificação de e-mails indesejados.

## Informações Fornecidas

Para resolver o problema, consideramos as seguintes probabilidades:

- \( P(S) \): Probabilidade de um e-mail ser spam.
- \( P(\sim S) \): Probabilidade de um e-mail **não** ser spam, que é \( 1 - P(S) \).
- \( P(E|S) \): Probabilidade de encontrar uma palavra suspeita em um e-mail que é spam.
- \( P(E|\sim S) \): Probabilidade de encontrar uma palavra suspeita em um e-mail que **não** é spam.

### Valores Dados

- \( P(S) = 0.4 \): A probabilidade de um e-mail ser spam.
- \( P(\sim S) = 0.6 \): A probabilidade de um e-mail não ser spam.
- \( P(E|S) = 0.8 \): A probabilidade de a palavra suspeita estar em um e-mail spam.
- \( P(E|\sim S) = 0.2 \): A probabilidade de a palavra suspeita estar em um e-mail não spam.

## Teorema de Bayes

Para calcular a probabilidade de um e-mail ser spam dado que uma palavra suspeita foi encontrada, usamos o Teorema de Bayes:

\[
P(S|E) = \frac{P(E|S) \cdot P(S)}{P(E)}
\]

Onde \( P(E) \) é a probabilidade total de encontrar a palavra suspeita, que pode ser calculada como:

\[
P(E) = P(E|S) \cdot P(S) + P(E|\sim S) \cdot P(\sim S)
\]

## Cálculo das Probabilidades

1. **Calcular \( P(E) \)**:
   - Usamos a fórmula acima para calcular a probabilidade total de encontrar a palavra suspeita.

2. **Aplicar o Teorema de Bayes**:
   - Usamos o valor de \( P(E) \) para calcular \( P(S|E) \), que representa a probabilidade de que um e-mail seja spam dado que a palavra suspeita foi encontrada.

## Resultado

A aplicação do Teorema de Bayes nos permite avaliar a eficácia do filtro de spam e sua precisão ao classificar e-mails com base em palavras suspeitas. Este exercício demonstra como a probabilidade condicional pode ser utilizada para melhorar a tomada de decisões em cenários incertos, como a filtragem de e-mails.


````python
P_spam = 0.60  # Probabilidade de ser spam
P_not_spam = 1 - P_spam  # Probabilidade de não ser spam (P(~Spam))
P_not_flagged_given_not_spam = 0.98  # Precisão para identificar e-mails não spam corretamente
P_flagged = 0.60  # Probabilidade de um e-mail ser sinalizado como spam

# Passo 1: Calcular P(Sinalizado | ~Spam)
P_flagged_given_not_spam = 1 - P_not_flagged_given_not_spam  # P(Sinalizado | ~Spam) = 0.02

# Passo 2: Usar o Teorema de Bayes para calcular P(~Spam | Sinalizado)
P_not_spam_given_flagged = (P_flagged_given_not_spam * P_not_spam) / P_flagged

# Quantidade de e-mails sinalizados na caixa de spam
emails_flagged_as_spam = 963

# Calcular o número de falsos positivos
false_positives = P_not_spam_given_flagged * emails_flagged_as_spam

# Exibir o resultado
print(f"Dos {emails_flagged_as_spam} e-mails na caixa de spam, cerca de {false_positives:.2f} são falsos positivos (não spam).")
````

# Explicação do Exercício 2.3

## Contexto do Exercício

O exercício envolve a utilização de uma cifra de César para criptografar e descriptografar um texto. A cifra de César é um método simples de substituição onde cada letra do texto é deslocada um número fixo de posições no alfabeto. Este exercício exemplifica como as cadeias de Markov podem ser aplicadas em processos de criptografia.

## Definição da Cifra de César

Na cifra de César, cada letra é substituída por outra letra que está um número fixo de posições à frente no alfabeto. Por exemplo, se o deslocamento for 3, a letra 'A' se tornará 'D', a letra 'B' se tornará 'E', e assim por diante.

### Deslocamento e Aplicação

Neste exercício, a cifra de César é aplicada com um deslocamento de 7. O enunciado menciona que esse deslocamento é aplicado três vezes. Portanto, o deslocamento total efetivo para a descriptografia será de \(7 \times 3 = 21\).

## Descriptografia

Para descriptografar o texto, precisamos reverter o deslocamento. Assim, aplicamos um deslocamento de -21 ao texto criptografado.

### Etapas do Processo de Descriptografia:

1. **Ler o texto criptografado**: O texto fornecido é "OCZ XJMMZXO VINRZM".
  
2. **Aplicar o Deslocamento Inverso**:
   - Para cada letra no texto, convertemos o caractere para seu valor ASCII.
   - Aplicamos o deslocamento de -21, ajustando as letras no alfabeto, considerando que o alfabeto contém 26 letras.

3. **Reconstituir o Texto Original**: Após aplicar o deslocamento, obtemos o texto descriptografado.

## Resultado

O resultado final da descriptografia será o texto original que foi codificado. Esse exercício demonstra a simplicidade e a eficácia da cifra de César, bem como o conceito de Markov, onde a saída depende unicamente do estado atual (ou letra) e não do histórico anterior. A cifra de César é um exemplo de um processo estocástico onde a probabilidade de cada letra é afetada pelo seu estado atual.


````python
# Função para descriptografar usando a cifra de César com deslocamento de -21
def caesar_cipher(text, shift):
    result = []
    for char in text:
        if char.isalpha():  # Apenas aplicar o deslocamento em letras
            shift_amount = shift % 26
            ascii_offset = 65 if char.isupper() else 97
            # Ajustar a posição da letra
            new_char = chr((ord(char) - ascii_offset + shift_amount) % 26 + ascii_offset)
            result.append(new_char)
        else:
            result.append(char)  # Manter espaços e caracteres especiais inalterados
    return ''.join(result)

# String criptografada
encrypted_text = "OCZ XJMMZXO VINRZM"

# Descriptografar aplicando a cifra de César com deslocamento -21
shift = -21
decrypted_text = caesar_cipher(encrypted_text, shift)

# Exibir o resultado
print(f"Texto descriptografado: {decrypted_text}")
````

# Explicação do Exercício 2.4

## Contexto do Exercício

O exercício envolve a criação e plotagem da função de massa de probabilidade (PMF) e da função de distribuição acumulada (CDF) para um dado de 6 lados. A PMF mostra a probabilidade de cada resultado individual, enquanto a CDF mostra a probabilidade acumulada até cada resultado.

## Definições

### Função de Massa de Probabilidade (PMF)

A PMF é uma função que dá a probabilidade de cada resultado possível em um experimento aleatório. Para um dado justo de 6 lados, a PMF é uniforme, pois cada face do dado tem a mesma probabilidade de ocorrer.

Para um dado de 6 lados, a PMF é:
- \( P(1) = \frac{1}{6} \)
- \( P(2) = \frac{1}{6} \)
- \( P(3) = \frac{1}{6} \)
- \( P(4) = \frac{1}{6} \)
- \( P(5) = \frac{1}{6} \)
- \( P(6) = \frac{1}{6} \)

### Função de Distribuição Acumulada (CDF)

A CDF é uma função que fornece a probabilidade de que uma variável aleatória \(X\) seja menor ou igual a um valor \(x\). Para o dado de 6 lados, a CDF é calculada somando as probabilidades da PMF.

Para o dado de 6 lados, a CDF é:
- \( P(X \leq 1) = P(1) \)
- \( P(X \leq 2) = P(1) + P(2) \)
- \( P(X \leq 3) = P(1) + P(2) + P(3) \)
- \( P(X \leq 4) = P(1) + P(2) + P(3) + P(4) \)
- \( P(X \leq 5) = P(1) + P(2) + P(3) + P(4) + P(5) \)
- \( P(X \leq 6) = P(1) + P(2) + P(3) + P(4) + P(5) + P(6) = 1 \)

## Plotagem

O exercício solicita a plotagem da PMF e da CDF:

1. **Plotagem da PMF**:
   - Um gráfico de barras é utilizado para mostrar a probabilidade de cada resultado ao lançar o dado.

2. **Plotagem da CDF**:
   - Um gráfico de linha é utilizado para mostrar a probabilidade acumulada. As áreas sob a curva da CDF também podem ser preenchidas para ilustrar melhor a soma das probabilidades.

## Resultado

Os gráficos gerados mostrarão visualmente a distribuição das probabilidades de resultados ao lançar um dado de 6 lados. A PMF fornecerá informações sobre a probabilidade de cada face do dado, enquanto a CDF mostrará como essas probabilidades se acumulam à medida que aumentamos os resultados.


````python
import numpy as np
import matplotlib.pyplot as plt

# Definir os resultados e suas respectivas probabilidades
outcomes = np.array([1, 2, 3, 4, 5, 6, 7, 8])
probabilities = np.array([0.10, 0.15, 0.20, 0.21, 0.17, 0.09, 0.06, 0.02])

# Calcular a distribuição acumulada (CDF)
cumulative_probabilities = np.cumsum(probabilities)

# Criar a função de massa de probabilidade (PMF)
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)  # Primeiro gráfico - PMF
plt.bar(outcomes, probabilities, color='blue', alpha=0.7)
plt.title("Função de Massa de Probabilidade (PMF)")
plt.xlabel("Resultado")
plt.ylabel("Probabilidade")

# Criar a função de distribuição acumulada (CDF)
plt.subplot(1, 2, 2)  # Segundo gráfico - CDF
plt.step(outcomes, cumulative_probabilities, where="mid", color='green', alpha=0.7)
plt.fill_between(outcomes, cumulative_probabilities, step="mid", alpha=0.3, color='green')
plt.title("Distribuição Acumulada de Probabilidade (CDF)")
plt.xlabel("Resultado")
plt.ylabel("Probabilidade Acumulada")

# Exibir os gráficos
plt.tight_layout()
plt.show()
````

# Explicação do Exercício 2.5

## Contexto do Exercício

O exercício envolve calcular a **entropia** de duas variáveis categóricas \(v_1\) e \(v_2\), além de medir a **informação mútua** entre elas. O objetivo é determinar quantos bits são necessários para codificar as variáveis independentemente e quanto podemos economizar ao codificar \(v_1\) sabendo \(v_2\).

### **Entropia**
A entropia, no contexto de teoria da informação, mede a quantidade de incerteza associada a uma variável. Quanto maior a entropia, mais bits são necessários para descrever a variável.

### **Informação Mútua**
A informação mútua quantifica a redução de incerteza sobre uma variável, dada a outra. No exercício, queremos saber quanto da entropia de \(v_1\) pode ser reduzida se soubermos \(v_2\), ou seja, o quanto de informação \(v_2\) nos fornece sobre \(v_1\).

## Passos para a Solução

### 1. **Definição dos Dados**

Os dados são fornecidos em formato de duas variáveis categóricas, \(v_1\) e \(v_2\), representando diferentes categorias.

### 2. **Cálculo das Probabilidades**

Para calcular a entropia, primeiro é necessário determinar as probabilidades de cada categoria em \(v_1\) e \(v_2\). Isso é feito calculando a frequência de cada categoria dividida pelo total de observações.

Para calcular as probabilidades de \(v_1\) e \(v_2\), usamos:

\[
P(v_1) = \frac{\text{número de ocorrências de cada valor de } v_1}{\text{total de observações}}
\]

\[
P(v_2) = \frac{\text{número de ocorrências de cada valor de } v_2}{\text{total de observações}}
\]

### 3. **Cálculo da Entropia**

A **entropia de Shannon** é calculada usando a seguinte fórmula:

\[
H(X) = -\sum p(x) \log_2 p(x)
\]

Onde:
- \( p(x) \) é a probabilidade de cada categoria em \(X\),
- \( H(X) \) é a entropia em bits.

No código, isso é calculado para as variáveis \(v_1\) e \(v_2\) usando a função `entropy()` da biblioteca **Scipy**:

```python
v1_info = entropy(v1_ps)  # Entropia de v1
v2_info = entropy(v2_ps)  # Entropia de v2


````python
import pandas as pd
from scipy.stats import entropy
from sklearn.metrics import mutual_info_score
# Load the data
data = {
    'v1': ['a', 'a', 'a', 'a', 'a', 'a', 'a', 'a',
           'b', 'b', 'b', 'b', 'b', 'b',
           'c', 'c', 'c', 'c', 'c', 'c', 'c',
           'd', 'd', 'd', 'd', 'd', 'd', 'd', 'd',
           'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e', 'e',
           'f', 'f', 'f', 'f', 'f', 'f'],
    'v2': ['a', 's', 'a', 'd', 'd', 'f', 'f', 'f',
           'f', 'g', 'g', 'g', 'g', 'h', 'h', 'h', 'h',
           'j', 'j', 'j', 'k', 'j', 'g', 'h', 'h', 'g',
           'f', 'f', 'f', 'd', 'd', 's', 's', 's',
           's', 's', 's', 'a', 'a', 'a', 'a', 'a', 'd',
           'f', 'g']
}
df = pd.DataFrame(data)
# The number of bits we need to encode v1 and v2 is their entropy. To calculate it, we need the probability of each possible outcome:
v1_ps = (df.groupby(by = "v1").size() / df.shape[0]).values
v2_ps = (df.groupby(by = "v2").size() / df.shape[0]).values

v1_info = entropy(v1_ps)
v2_info = entropy(v2_ps)

print(v1_info)
print(v2_info)
mi = mutual_info_score(df["v1"], df["v2"])
print("Knowing v2 gives us %1.2f%% of v1" % (100 * mi / v1_info))
````

#Exercício 4.2

````python
import networkx as nx
from collections import defaultdict

# Definindo os dados manualmente como uma lista de arestas com informações de tipo
edge_list = [
    (1, 2, "C"),
    (1, 3, "C"),
    (1, 4, "C"),
    (5, 6, "C"),
    (5, 7, "C"),
    (5, 8, "C"),
    (9, 10, "C"),
    (9, 11, "C"),
    (9, 12, "C"),
    (13, 14, "C"),
    (13, 15, "C"),
    (13, 16, "C"),
    (1, 5, 1),
    (1, 9, 1),
    (1, 13, 1),
    (5, 9, 1),
    (5, 13, 1),
    (9, 13, 1),
    (2, 6, 2),
    (6, 10, 2),
    (10, 14, 2),
    (14, 2, 2),
    (3, 7, 3),
    (7, 11, 3),
    (11, 15, 3),
    (4, 8, 4),
    (12, 16, 4)
]

# Criando o multigrafo a partir da lista de arestas
G = nx.MultiGraph()

# Adicionando as arestas ao grafo
for u, v, edgetype in edge_list:
    G.add_edge(u, v, edgetype=edgetype)

# Determinar quantas camadas temos
layers = len(set(e[2]["edgetype"] for e in G.edges(data=True))) - 1

# Se isso for um acoplamento em cadeia, todos os nós podem ter no máximo duas arestas de acoplamento
node_cedges = defaultdict(int)
for e in G.edges(data=True):
    if e[2]["edgetype"] == "C":
        node_cedges[e[0]] += 1
        node_cedges[e[1]] += 1

if max(node_cedges.values()) == 2:
    print("Chain coupling")

# Se isso for um acoplamento de clique, todos os nós devem ter mais de uma aresta de acoplamento
if min(node_cedges.values()) > 1:
    print("Clique coupling")

# Se isso for um acoplamento em estrela, todos os nós devem ter uma aresta ou |L| - 1 arestas de acoplamento
if set([1, layers - 1]) == set(node_cedges.values()):
    print("Star coupling")
````

#Exercício 4.3

````python
import networkx as nx
from itertools import combinations
hyperedges = []
with open("data.txt", 'r') as f:
   for line in f:
      hyperedges.append(line.strip().split())
G_uni = nx.Graph()
for hyperedge in hyperedges:
   for combination in combinations(hyperedge, 2):
      G_uni.add_edge(*combination)
G_bip = nx.Graph()
hyperid = 1
for hyperedge in hyperedges:
   for node in hyperedge:
      G_bip.add_edge(node, "h%s" % hyperid)
   hyperid += 1
````

#Exercício 4.4

````python
import networkx as nx
G = nx.read_edgelist("data.txt", create_using = nx.MultiGraph(), data = [("start", int), ("end", int)])
first_window = nx.Graph()
second_window = nx.Graph()
third_window = nx.Graph()
for e in G.edges(data = True):
   if e[2]["start"] <= 3:
      first_window.add_edge(e[0], e[1])
   if e[2]["start"] <= 6 and e[2]["end"] > 3:
      second_window.add_edge(e[0], e[1])
   if e[2]["end"] > 6:
      third_window.add_edge(e[0], e[1])
````
